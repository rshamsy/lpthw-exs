```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
#displays plots inside jupyter notebook
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("/Users/MHK/Desktop/Pokemon.csv",encoding = "ISO-8859-1",index_col=0)
```

```python
df.head()
```

###### Scatter Plot

```python
sns.lmplot(x='Attack',y='Defense',data = df)
```

```python
sns.lmplot(x='Attack',y='Defense',data=df,fit_reg=False,hue='Stage')

plt.axis([0,None,0,None])
```

###### Matplotlib Pyplot Tutorial

```python
plt.plot([1,2,3,4])
plt.ylabel("some numbers")
plt.xlabel('data count')
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.xlabel('x values')
plt.ylabel('y values')
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16],'ro')
plt.axis([0,6,0,20])
plt.xlabel('x values')
plt.ylabel('y values')
```

```python
t = np.arange(1.,10.,0.5)
plt.plot(t,t,'r--',t,t**2,'bo',t,t**3,'g^')
```

```python
df_reduced = df.drop(labels=['Total','Stage','Legendary'],axis=1)
df_reduced.head()
```

```python
sns.boxplot(data=df_reduced)
```

```python
plt.figure(figsize=[10,10])
sns.set_style('whitegrid')
sns.violinplot(x='Type 1',y='Attack',data=df_reduced)
```

```python
help(sns.swarmplot)
```

```python
help(plt.figure)
```

```python
plt.figure(figsize=[10,10])
sns.swarmplot(x='Type 1', y='Attack', data = df_reduced,split=True)
```

```python
plt.figure(figsize=(10,10))
sns.violinplot(x='Type 1',y='Attack',data=df_reduced,inner=None)
sns.swarmplot(x='Type 1', y='Attack',data=df_reduced,split=True,color='b')
plt.ylim(-50,250)
```

```python
sns.set_style('whitegrid')
corr = df_reduced.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr)
```
